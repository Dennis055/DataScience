```python
import pandas as pd
from bs4 import BeautifulSoup
import json
import requests
import re
```

```python
keyword = '除濕機'
pages = 30
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
urls = []
for page in range(1, pages):
    url = 'https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage={}&searchType=1&cateLevel=2&ent=k&searchKeyword={}&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType'.format(page, keyword)
    print(url)
    resp = requests.get(url, headers=headers)
    if resp.status_code == 200:
        soup = BeautifulSoup(resp.text)
        for item in soup.select('li.goodsItemLi > a'):
            urls.append('https://m.momoshop.com.tw'+item['href'])
    urls = list(set(urls))
    print(len(urls))
#     break
```

```python
df = []
for url in urls:
    print(url)
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.text)
    break
soup
```

```python
# title
print(soup.find('meta',{'property':'og:title'})['content'])

#品牌
print(soup.find('meta',{'property':'product:brand'})['content'])

# 連結
print(soup.find('meta',{'property':'og:url'})['content'])

# 原價
print(re.sub(r'\r\n| ','',soup.find('del').text))
# 特價
print(soup.find('meta',{'property':'product:price:amount'})['content'])

# 幣別
print(soup.find('meta',{'property':'product:price:currency'})['content'])

# 類型
cate = ''.join([i.text for i in soup.findAll('article',{'class':'pathArea'})])
cate = re.sub('\n|\xa0',' ',cate)
print(cate)

# 描述
print(re.sub(r' |\n|\r','',soup.findAll('div',{'class':'Area302'})[1].text))

# 詳情
print(re.sub(r' |\n|\r','',soup.find('div',{'id':'_div202html'}).text))
```
